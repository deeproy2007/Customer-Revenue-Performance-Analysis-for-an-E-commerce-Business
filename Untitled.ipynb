```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Ecommerce.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Product_Category'],inplace=True)
```

```python
df.describe(include='all')
```

```python
df['Gender'].replace({'male':'Male','female':'Female'},inplace=True)
df['Age']=df['Age'].astype(int)
df['Quantity']=df['Quantity'].astype(int)
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date'] = df['Date'].dt.strftime('%m-%d-%y')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df['Date'] = pd.to_datetime(df['Date'])
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Total_Amount'].sum()

monthly_sales.plot(title='Monthly Revenue Trend')
```

```python
category_sales = df.groupby('Product_Category')['Total_Amount'].sum().sort_values(ascending=False)
sns.barplot(x=category_sales.values,hue=category_sales.index,palette='cividis')
plt.title('Product category wise sales')
plt.show()
```

```python
df[['Discount_Amount', 'Total_Amount']].corr()
```

```python
a=df.groupby('City')['Total_Amount'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=a.values,hue=a.index,palette='deep')
plt.title('Top City')
plt.show()
```

```python
# customer_value = 
df.groupby('Customer_ID')['Total_Amount'].sum()


```

```python
plt.figure(figsize=(20,6))
plt.subplot(1,2,1)
b=df.groupby('City')['Customer_Rating'].mean().sort_values(ascending=False)
plt.title('City wise customer rating')
plt.xlabel('Rating')
sns.barplot(x=b.values,hue=b.index,palette='viridis')
c=df.groupby('Delivery_Time_Days')['Customer_Rating'].mean()
plt.subplot(1,2,2)
sns.lineplot(x=c.index,y=c.values)
plt.title('Customer rating wise Delivery time trend')
plt.ylabel('Rating')
plt.show()
```

```python
df['Payment_Method'].value_counts(normalize=True).plot(kind='pie',autopct='%.2f%%',startangle=90)
plt.title('Patment Method wise Analysis')
```

```python
mycolor=['orange','green']
df.groupby('Is_Returning_Customer')['Total_Amount'].mean().plot(kind='pie',autopct="%.2f%%",startangle=90,colors=mycolor)
font1={'color':'red','size':16}
plt.title("Repeted customer OR not",fontdict=font1)
```

```python
e=df.groupby('Device_Type')['Total_Amount'].mean()
sns.barplot(x=e.index,y=e.values,palette='viridis')
plt.title('Device_type wise Avg sales')
plt.ylabel('Avg sales')
plt.show()
```
